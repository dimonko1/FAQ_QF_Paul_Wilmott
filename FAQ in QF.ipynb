### <font color=7E87FC>2.2 What is arbitrage?</font>
#### • SHORT ANSWER: making a sure profit in excess of the risk-free rate. It is a portfolio of zero value today and which is of positive value in the future with positive probability and of negative value in the future with zero probability.

### <font color=7E87FC>2.3 What is Put-Call Parity?</font>
#### • SHORT ANSWER: put-call parity – relationship between European call and European put with the same underlying, stock, expiration:
#### $$c - p = S - Ke^{-r(T-t)}$$
#### If the stock is above the strike at expiration we will have $S-K$ from call, $0$ from put and $-S$ from stock. That's total of $-K$ ($T-T=0$, and the exponent vanishes). If the stock is below strike, call gives us $0$, put gives $-(K-S)$, and total $-K$. In both cases it sums to zero, because the amount $K$ is discounted. That's <font color=7E87FC>model-independent</font> relationship.
#### In put-call parity options will also have similar implied volatility.
#### In practice due to a bid/offer spread that's not ideal. If we're looking at the violations of PC-parity, then we must use bid (offer) if we're going short(long) the options.
#### In terms of implied volatility, it's much easier to spot violations of PC-parity. Suppose, that the bid/offer on a call is $22\%-25\%$ in implied volatility terms and that on a put (same strike and expiration) is $21\%-23\%$ – there's overlapping range $22\%-23\%$ – no arbitrage opportunities. If the put prices were $19\%-21\%$, then that would be violation of PC-parity. But in practice that almost doesn't happen.
#### If we have dividends during the life of the options (assuming cashflows are deterministic):
#### $$c - p = S - \text{PV of dividends} - Ke^{-r(T-t)}$$
#### If interest rates are not constant, then just use $0$-coupon bonds to discount strike and dividends.
#### PC-parity doesn't work with American options, because the short exposition could be exercised against us, leaving us with some exposure to the stock price.

### <font color=7E87FC>2.4 What is the Central Limit Theorem and What are its Implications for Finance?</font>
#### • SHORT ANSWER: the distribution of the average of lots of r.vs. will be normal (Gaussian) even when the individual numbers are not normally distributed. R.vs. should be drawn from the same distribution.
#### • EXAMPLE: roll the dice many times.
#### • LONG ANSWER: let $X_1,X_2,\ldots,X_n$ be a sequence of r.vs. which are i.i.d. with finite mean $m$ and std. dev. $s$. The sum:
#### $$S_n = \sum_{i=1}^n X_i$$ 
#### with mean $mn$ and std. dev. $s\sqrt{n}$. As $n$ gets larger, the distribution of $S_n$ tends to the Normal distribution. We can make it standard Normal:
#### $$\overline{S_n}=\frac{S_n - mn}{s\sqrt{n}}$$
#### Std. dev. of a simple coin-tossing experiment grows less than the expectation (square-root rule).
#### Application in finance – HFT and normal distribution of stock/index/currency/commodity returns but we shall be careful! As long as none of the random variables has too much more impact on the average than the others then it still works (in case distributions are different). Also there might exist dependence between the r.vs.
#### If the tails of the individual distributions have a power-law decay, $|x|^{-1-\alpha}$ with $0<\alpha<2$, then average will tend to a stable Lévy distribution.
#### If we multiply i.i.d. r.vs then we converge to the <font color=7E87FC>lognormal distribution</font> (stock prices themselves).

### <font color=7E87FC>2.7 What is Extreme Value Theory?</font>
#### • SHORT ANSWER: mathematics behind extreme events. CLT deals with averages, EVT – with  extremes.
#### • LONG ANSWER: throw Normal distribution away, its tails are too thin to capture market crashes.
#### Important EVT result concerns the distribution of maxima and minima.
#### If $X_i$ are i.i.d. r.vs. and $x = \text{max}(X_1,X_2,\ldots,X_n)$, and the distribution of $x$ converges to
#### $$\frac{1}{\sigma}(1 + \xi\frac{\xi(x-\mu)}{\sigma})^{-1/\xi-1}\exp(-(1+\frac{\xi(x-\mu)}{\sigma})^{-1/\xi})$$
#### If $\xi=0$ we get a Gumbel distribution, $\xi<0$ – a Weibull, $\xi > 0$ – a Frechet (most used in finance).

### <font color=7E87FC>2.8 What is CrashMetrics?</font>
#### • SHORT ANSWER: CrashMetrics – a stress-testing methodology for evaluating portfolio performance in the even of extreme movements in financial markets.
#### • LONG ANSWER: CrashMetrics is used to examine the effects of a large move in the market as a whole / to study times when diversification doesn't work. 
#### Let's say our portfolio consists of single underlying stock and its derivatives:
#### $$\delta \Pi = F(\delta S)$$
#### In CrashMetrics the risk is measured as the worst case over some range of stock movement:
#### $$\text{Worst-cae loss} = \underset{-\delta S^- ≤ \delta S ≤ \delta S^+}{\text{min}} F(\delta S)$$
#### This is the number that would be quoted as the possible downside during a dramatic move in the market. We can reduce it by adding derivatives – <font color=7E87FC>Platinum Hedging</font>. E.g. use some OTM puts to make the worst case not so that:
#### $$F(\delta S)+\lambda F^*(\delta S)-|\lambda|C$$
#### $F^*$ – formula for the change in value of the hedging contract, $C$ – cost of each hedging contract, $\lambda$ – quantity of contracts.
#### We can relate the return on $i$th asset to the return on a representative index, where $\kappa_i$ – <font color=7E87FC>crash coefficient</font>:
#### $$\frac{\delta S_i}{S_i} = \kappa_ix$$
#### If $\kappa = 1.2$, then the fall of $10\%$ in index will be met a fall of $12\%$ in the stock.
#### Therefore, we consider the worst case of:
#### $$\delta \Pi = F(\delta S_1, \ldots, \delta S_N) = F(\kappa_1xS_1, \ldots, \kappa_NxS_N)$$
#### as our measure of downside risk.
#### $$\delta \Pi = F(\kappa_1xS_1, \ldots, \kappa_NxS_N) + \sum_{k=1}^M \lambda_kF_k(\kappa_1xS_1,\ldots, \kappa_NxS_N) - \sum_{k=1}^M|\lambda_k|C_k$$
#### where $F$ – original portfolio, $F_k$ are the $M$ available hedging contracts.

### <font color=7E87FC>2.9 What is a Coherent Risk Measure and What are its Properties?</font>
#### • SHORT ANSWER: a risk measure is coherent if it satisfies certain simple, mathematical propertie: sub-additivity (adding two risky portfolios cannot increase the measure of risk) = merger doesn't not create extra risk.
#### • LONG ANSWER:
#### 1. Sub-additivity: $\rho(X+Y)≤\rho(X)+\rho(Y)$ – if you add two portfolios together the total risk can't get any worst ean adding the two risks separately. $≤$ becuase the effect of scale of cancellation effects may make the risk better.
#### 2. Monotonicity: if $X≤Y$ for each scenario then $\rho(X)≥\rho(Y)$ – if one portfolio has better values than another under all scenarios then its risk will be better.
#### 3. Positive homogenity: for all $\lambda >0$, $\rho(\lambda X) = \lambda \rho(X)$ – double your portfolio then you double your risk.
#### Translation invariance: for all $c=\text{const}$, $\rho(X+c)=\rho(X)-c$ – adding cash to your portfolio lowers the risk.
#### Risk measure that satisfies all of these is coherent. VaR is not coherent! A popular corehent measure is <font color=7E87FC>Expected Shortfall</font> – the average of all the P&Ls making up the tail percentile of interest.

### <font color=7E87FC>2.10 What is Modern Portfolio Theory?</font>
#### • SHORT ANSWER: the MPT of Harry Markowitz (1952) introduced the analysis of portfolios of investments by considering the expected return and risk of individual assets and their interrelationship as measure by correlation. Diversification plays an important role.
#### • LONG ANSWER: the returns are assumed to follow Normal distribution. Markowitz tells us how to consider the correlation between the number of stocks to diversifty effectively.
#### In MPT world of $N$ assets there are $2N+N(N-1)/2$ parameters: expected return, std. dev, correlations. Let's say $\mu_A$ – expected return from portfolio $A$, $\sigma_A$ – std. dev. of $A$ etc. Portfolio $A$ is surely better than $B$ if:
#### $$\mu_A ≥ \mu_B \text{ and }\sigma_A≤\sigma_B$$ $$\mu_{\Pi}=\sum_{i=1}^N W_i\mu_i$$ $$\sigma_{\Pi} = \sqrt{\sum_{i=1}^N \sum_{j=1}^N W_iW_j\rho_{ij}\sigma_i\sigma_j}$$
#### The curve is risk-return space is the <font color=7E87FC>efficient frontier</font>. If we add the risk-free asset, then we get <font color=7E87FC>Capital Market Line</font>, and the point at which it is tangential – <font color=7E87FC>Market Portfolio</font>.

### <font color=7E87FC>2.11 What is the Capital Asset Pricing Model?</font>
#### • SHORT ANSWER: CAPM introduces <font color=7E87FC>specific risk</font> (unique to an individual asset), <font color=7E87FC>systematic risk</font> (associated with the market). Investors are compensated for taking systematic risk, not the specific one (it's diversifiable).
#### • EXAMPLE: stock has $\mu=0.15$, $\sigma = 0.2$, but how much of that risk and return are related to the market as a whole? The less they are, the more the stock will be diversifiable.
#### $R_i$ – return of $i$-th asset, $R_M$ – representative index:
#### $$R_i = \alpha_i + \beta_i R_M + \varepsilon_i$$
#### Expected return on the $i$-th asset:
#### $$\mu_i = \alpha_i + \beta_i\mu_M $$ $$\sigma_i=\sqrt{\beta_i^2\sigma_M^2 + e_i^2}$$
#### If we have portfolio of such assets, then the return is given by:
#### $$\frac{\delta \Pi}{\Pi}= \sum_{i=1}^N W_iR_i = (\sum_{i=1}^N W_i\alpha_i) + R_M (\sum_{i=1}^N W_i\beta_i) + \sum_{i=1}^N W_i\varepsilon_i$$ $$\mu_{\Pi} = (\sum_{i=1}^N W_i\alpha_i) + \mathbb{E}[R_M](\sum_{i=1}^N W_i\beta_i)$$ $$\alpha_{\Pi}=\sum_{i=1}^N W_i\alpha_i$$ $$\beta_{\Pi}=\sum_{i=1}^NW_i\beta_i$$ $$\mu_{\Pi} = \alpha_{\Pi}+\beta_{\Pi}\mathbb{E}[R_M]=\alpha_{\Pi}+\beta_{\Pi}\mu_M$$
#### The risk in $\Pi$ is:
#### $$\sigma_{\Pi} = \sqrt{\sum_{i=1}^N\sum_{j=1}^N W_iW_j\beta_i\beta_j\sigma_M^2 + \sum_{i=1}^NW_i^2e_i^2}$$

### <font color=7E87FC>2.12 What is Arbitrage Pricing Theory?</font>
#### • SHORT ANSWER: the APT by Stephen Ross (1976) represents the returns on individual assets as a lienar combination of multiple random factors.
#### • LONG ANSWER: the MPT represents each asset by its random return and then links the returns on different assets via a correlation matrix. The CAPM returns on individual assets are related to returns on the market as a whole together with an uncorrelated stock-specific random component. The APT treats returns as a linear combination of multiple random factors with proper weights.
#### Random return on the $i$-th asset:
#### $$R_i = \alpha_i + \sum_{j=1}^n \beta_{ji}\overline{R_j}+\varepsilon_i$$ $$\sum_{i=1}^Na_iR_i = \sum_{i=1}^Na_i\alpha_i + \sum_{j=1}^n(\sum_{i=1}^Na_i\beta_{ij})\overline{R_j}+\ldots$$
#### Factors must be sufficient to represent the economy.

### <font color=7E87FC>2.13 What is Maximum Likelihood Estimation?</font>
#### • SHORT ANSWER: MLE – statistical technique for estimating parameters in a probability distribution. We choose parameters that maximize <font color=7E87FC>a priori</font> probability.
#### • LONG ANSWER: we have three hats containing normally distributed r.vs. Hat A $\sim N(0,0.1^2)$, hat B $\sim N(0, 1)$, hat C $\sim N(0, 10^2)$.
#### We pick randomly and get $-2.6$.
#### Pick from A: $\frac{1}{\sqrt{2\pi}0.1}\exp(-\frac{2.6^2}{2\cdot 0.1^2}) = 6 \cdot 10^{-147}.$ Very unlikely!
#### Pick from B: $\frac{1}{\sqrt{2\pi}1}\exp(-\frac{2.6^2}{2\cdot 1^2}=0.014$
#### Pick form C: $\frac{1}{\sqrt{2\pi}10}\exp(-\frac{2.6^2}{2\cdot 10^2}=0.039$
#### Pick from C is the most likley one with the hightest probability for picking the number $-2.6$.
#### <font color=7E87FC>Find the volatility</font>. We have one hat $\sim N(0,\sigma)$, $\sigma$ is unknown. We draw $N$ numbers $\phi_i$ from this hat. Estimate $\sigma$.
#### What is the probability of drawing $\phi_i$ from this distribution? It is: $\frac{1}{\sqrt{2\pi}\sigma}\exp(-\frac{\phi_i^2}{2\sigma^2}).$
#### What is the probability of drawing all numbers $\phi_1, \phi_2, \ldots, \phi_N$ from indepdent hat? $\prod_{i=1}^N\frac{1}{\sqrt{2\pi}\sigma}\exp(-\frac{\phi_i^2}{2\sigma^2}).$
#### We maximize $\sigma$. First take logarithms of this expression, and then diffrerentiate w.r.t. $\sigma$ and set it to $0$: 
#### $$\frac{d}{d\sigma}(-N\ln(\sigma)-\frac{1}{2\sigma^2}\sum_{i=1}^N\phi_i^2)=0$$ $$-\frac{N}{\sigma}+\frac{1}{\sigma^3}\sum_{i=1}^N\phi_i^2 = 0$$ $$\sigma^2 = \frac{1}{N}\sum_{i=1}^N\phi_i^2$$
#### We should recognize this as a measure of the variance.

### <font color=7E87FC>2.14 What is Cointegration?</font>
#### • SHORT ANSWER: two TSs are cointegrated if a lienar combination has constant mean and std.dev. Simply speaking, they never stray too far from one another. It's the basiccs for <font color=7E87FC>pairs trading</font>.
#### • LONG ANSWER: cointagration is a more robust measure of the linkage between two financial quantities but the derivatives theory doesn't use this concept a lot (e.g. while hedging we are more interested in short scales, where the correlation rules).
#### Two stocks may be perfectly correlated in the short-run yet they diverge in the long-run.
#### In portfolio theory, even though one stock price path may not be stationary, a linear combination of assets (portoflio) may be! Can we find $\lambda_i$ with $\sum_{i=1}^N\lambda_i = 1$ such that $\sum_{i=1}^N \lambda_iS_i$ is stationary? If yes, the stocks are cointegrated. It may be useful in <font color=7E87FC>tracking the index</font> (e.g. use 15 stocks to track the S&P 500).
#### Another feature of cointegration is <font color=7E87FC>Granger causaility</font> which is where one variable leads and another lags – this explains why there is any dynamic relationship between several financial quantities.

### <font color=7E87FC>2.15 What is the Kelly Criterion?</font>
#### • SHORT ANSWER: The Kelly Criterin is a technique for maximizing expected growth of assets by optimally investing a fixed fraction of your wealth in a series of investments.
#### • EXAMPLE: we have a biased coin with probability of heads $p>0.5$. We find someone willing to bet any amount against us at evens any number of times. We start with 1000 USD. How much of this should we bet?
#### • LONG ANSWER: what proportion to bet is the Kelly criterion.
#### Let's denote r.vs. taking 1 value with $p, (1-p)$ by $\phi$ and $f$ to denote the fraction of our wealth to bet. The growth of wealth after each toss of the coin is thus the r.v.:
#### $$\ln(1+f\phi)$$
#### Expected growth rate is
#### $$p\ln(1+f) + (1-p)\ln(1-f)$$
<center><img src=https://sun9-17.userapi.com/impg/CMWneLjdjvCiL2Jp4TUN2JhtWsYh5S5KybnJrg/oJNtueEhbss.jpg?size=1280x992&quality=96&sign=0069913f90ace5cd21f15dcd1e3f1760&type=album width=350></img></center>

#### Expected growth rate if maximized by the choice:
#### $$f = 2p-1$$ 
#### This is the <font color=7E87FC>Kelly fraction</font>. Investing less the that will be a conservative strategy; investing more will result in adding volatility to returns and decreasing the expected returns.
#### More generally, if the investment has an expected return of $\mu$ and $\sigma \gg \mu$ then the expected growth for an investment of fraction $f$ is:
#### $$\mathbb{E}[\ln(1+f\phi)]=f\phi - \frac{1}{2}f^2 \phi^2 + \ldots$$
#### The Kelly fraction = maximizing this expression:
#### $$f = \frac{\mu}{\sigma^2}$$
#### IRL these parameters are rarely known accurately, so one would make a smaller bet – <font color=7E87FC>half Kelly</font>.

### <font color=7E87FC>2.16 Why Hedge?</font>
#### • SHORT ANSWER: hedging means the reduction of risk by exploiting relationships or correlation (lack of it) between various risky investments. It results in an improved risk/return.
#### • LONG ANSWER: two main classifications:
#### – <font color=7E87FC>model-independent hedging</font> – e.g. put-call parity. This hedging is independent of how the underlying asset changes in value. Another example is spot-forward parity;
#### – <font color=7E87FC>model-dependent hedging</font> – most sophisticated ones. Depend on the model for the underlying. E.g. Black-Scholes.
#### <font color=7E87FC>$\Delta$-heding</font> – perfect elimination of all risk by using correlation between the option value and the underlying – <font color=7E87FC>dynamic hedging</font> (transaction costs may kill it). If assets depend on the same risk, we can delta hedge them inversly proportional to these deltas (one long, one short). But solo delta hedging has errors due to its convexity...
#### <font color=7E87FC>$\Gamma$-hedging</font> (exploiting convexity) is used to reduce the size of each rehedge and/or increase the time between rehedging and thus costs.
#### <font color=7E87FC>$\nu$-hedging</font> (vega) – we shouldn't use constant vol. models (e.g. BS-model) to calculate sensitivities to parameters that are assumed not to vary. Variables (underlying, time). parameters (vol., div. yield, interest rate).
#### <font color=7E87FC>Static hedging</font> – set up a hedge and leave it.
#### <font color=7E87FC>Superhedging</font>.
#### <font color=7E87FC>Margin hedging</font> – one of the problems is the requirement to suddenly come up with a large amount of cash to cover an unexpected margin call – this often causes banks and other fin. institutions to suffer during volatile markets. Examples – Metallgesellschaft and LTCM. Not a problem on OTC markets.

### <font color=7E87FC>2.17 What is Marking to Market and How Does it Affect Risk Management in Derivatives Trading?</font>
#### • SHORT ANSWER: Marking to market (MtM) means valuing an instrument at the price at which it's currently trading in the market. 
#### • EXAMPLE: if the current market price of a stock is 47 USD, but we think it's undervalued and it should cost 60 USD, then saying our portfolio costs 47 is MtM, whilst saying 60 – marking to (our) model.
#### • LONG ANSWER: MtM of exchange-traded instruments is very straightforward – just look at the last trading price. The real problem happens with the OTC instruments, because they aren't traded actively, they may be unique to us and our counterparties. Such instruments have to be <font color=7E87FC>market to model</font>, and the main question here is WHAT VOLATILITY TO USE?
#### Some possible ways of marking OTC contracts:
##### – the trader uses his own volatility, usually close to the IV;
##### – use prices obtained from brokers (real, tradeable, unprejudiced). The drawback is that you can't call your broker forever for prices with no intention of trading;
##### – use a vol. model that is calibrated to vanillas.
#### The option, being exchange traded, would probably be marked to market using the IV, but the ultimate profit will depend on the realized volatility and also how the option is hedged.

### <font color=7E87FC>2.18 What is the Efficient Markets Hypothesis?</font>
#### • SHORT ANSWER: EMH means it's impossible to beat the market because all information about securities is already reflected in their prices.
#### "I’d be a bum in the street with a tin cup if the markets were efficient," – Warren Buffett.
#### • LONG ANSWER: the concept of EMH was proposed by Eugene Fama in the 1960s. There are a few forms of EMH:
##### – Weak-form efficiency – returns can't be made by using investment strategies based on historical prices or other historical financial data. This means technical analysis sucks!
##### – Semi-strong form efficiency – a trading strategy incorporating current publicly available fundumanetal information (such as fin. statement) and historical price information will not systematically outperform a buy-and-hold strategy – fundamental analysis sucks!
##### – Strong-form efficiency – share prices reflect all information, public and private, fundamental and historical, and no one can earn excess returns – inside information sucks!
#### A weaker version of EHM is <font color=7E87FC>the Adaptive Market Hypothesis</font> of Andrew Lo – even if the profitable strategy or opportunity appears, others will find out about it.
#### Many studies claim that the weak form of EMH quite works.

### <font color=7E87FC>2.19 What are the Most Useful Performance Measures?</font>
#### • SHORT ANSWER: used to quantify the results of a trading strategy, usually adjusted for risk.
#### • EXAMPLE: stock's average growth is 10% per annum, another one's is 30% per annum. At a first glance the second one is more preferable, but let's add some details. The first stock's vol. is 5%, whereas the second was 20% – does that make difference?
#### • LONG ANSWER:
#### – <font color=7E87FC>Sharpe ratio</font>:
#### $$\text{Sharpe} = \frac{\mu - r}{\sigma}$$
#### It's quoted in annualized terms. High Sharpe ratio is intended to be a sign of a good strategy. If returns are distributed normally, then the ratio is related to the probability of making a return in excess of the risk-free rate. 
#### In MPT the Sharpet ratio is the slope of the line joining each investment to the risk-free investment. Choosing the portfolio that maximizes the Sharpe ratio gives us the <font color=7E87FC>Market Portfolio</font>.
#### The Sharpe ratio has been criticized for attaching equal weight to upside ‘risk’ as downside risk since the standard deviation incorporates both in its calculation. This may be important if returns are very skewed.
#### – <font color=7E87FC>Modigliani - Modigliani measure</font> (M2) is a simple linear transformation of the Sharpe ratio:
#### $$\text{M2} = r + v\times \text{Sharpe}$$
#### $v$ – std. dev. of returns of the relevant benchmark – interpreted as the return we would expect from our portfolio if it were (de)leveraged to have the same volatility as the benchmark.
#### – <font color=7E87FC>Sortino ratio</font> is the same as the Sharpe ratio, except it uses the $\sqrt{\text{ }}$ of the semi-variance as the denominator measuring risk. Semi-variance is measure in the same way as the variance, except that all the data poitns with positive return are replaced with zero or some target value.
#### – <font color=7E87FC>Treumpr ratio</font> or Reward-to-variability ratio is another Sharpe-like measure, but the denominator is the systematic risk ($\beta$ from CAPM):
#### $$\text{Treynor} = \frac{\mu - r}{\beta}$$
#### In a well-diversified portfolio Sharpe and Treynor are similar. Treynor is mostly used with individual stocks or less-diversified portfolios.
#### – <font color=7E87FC>Information ratio</font> uses the idea of tracking error. Tracking error = std. dev. of the differences between the portfolio returns and the benchmark returns. It indicated the valued added by a manager relative to their benchmark.
#### $$\text{Information ratio} = \frac{\mu - r}{\text{Tracking error}}$$

### <font color=7E87FC>2.20 What is a Utility Function and How is it Used?</font>
#### • SHORT ANSWER: it represents the "worth", "happiness" or "satisfaction" associated with goods, services, events, levels of wealth etc. 
#### • LONG ANSWER: rational people are supposed to act so as to increase their utility.
#### If we can compre utility numerically, we're dealing with <font color=7E87FC>cardinal utility</font>. If the ordering is all that matters, then it's <font color=7E87FC>ordinal utility</font>.
#### Denote utility function by $U(W)$, where $W$ is the wealth. The function $U(W)$ can vary among investors, each will have a different attitude to risk.
#### $U'(W)≥0$ – more is preferred to less. This indicates the marginal improvement in utility with changes in wealth.
#### Usually $U''(W)<0$ – the utility function is strictly concave. It's the rate of change of the marginal "happiness", it get's harder and harder to increase happiness as wealth increases. An investor is said to be <font color=7E87FC>risk averse</font> if his utility function is concave – the law of diminishing returns.
#### The <font color=7E87FC>absolute risk aversion function</font> is:
#### $$A(W) = -\frac{U''(W)}{U'(W))}$$
#### The <font color=7E87FC>relative risk aversion</font> is:
#### $$R(W) = -\frac{WU''(W)}{U'(W)}=WA(W)$$
#### When the wealth is random, and all outcomes can be assigned a probability, one can ask what amount of certain wealth has the same utility as the expected utility of the unknown outcomes. Simply solve:
#### $$U(W_c) = \mathbb{E}[U(W)]$$
#### The quantity $W_c$ that solves this equation is the <font color=7E87FC>certainty equivalent wealth</font>.

### <font color=7E87FC>2.22 What is a Wiener Process/Brownian Motion and What are its Uses in Finance?</font>
#### • SHORT ANSWER: it's a stochastic process with stationary independent normally distributed increments and which also has continuous sample paths. It's the most common stochastic building block for random walks in finance.
#### • LONG ANSWER: mathematically, BM is a continuous, stationary, stochastic process with independent normally distributed increments. If $W_t$ – the BM at time $t$, then for every $t$, $\tau ≥ 0, W_{t+\tau} - W_t$ is independent of $\{W_u:0≤u≤t\}$ and has a normal distribution with zero mean and variance $\tau$.
#### Important properties of BM:
##### – finiteness: the scaling of the variance with the time step is crucial to BM remaining finite;
##### – continuity: the paths are continuous, the path is fractal and not differentiable anywhere;
##### – Markov: the conditional distribution of $W_t$ given information up until $\tau<t$ depends only on $W_{\tau}$;
##### – Martingale: given information up until $\tau<t$ the conditional expectation of $W_t$ is $W_{\tau}$;
##### – Quadratic variation: if we divide up the time $0$ to $t$ in a partition with $n+1$ partition points $t_i=it/n$, then
##### $$\sum_{j=1}^n(W_{t_j}-W_{t_{j-1}})^2 \rightarrow t$$
##### – normality: over finite time increments $t_{i-1}$ to $t_i$, $W_{t_i} - W_{t_{i-1}}$  is normally distributed with mean zero and variance $t_i - t_{i-1}$.
#### $dW$ is OK to be thought of as a random number drawn from a normal distribution with $\mathbb{E}[dW] = 0, \mathbb{E}[dW^2]=dt$.

### <font color=7E87FC>2.23 What is Jensen's Inequality and What is its Role in Finance?</font>
#### • SHORT ANSWER: Jensen's inequality states that if $f(\dot)$ is a convex function and $x$ is a r.v., then 
#### $$\mathbb{E}[f(x)]≥f(\mathbb{E}[x])$$
#### This justifies why non-linear instruments, options, have inherent value.
#### • EXAMPLE: roll a die, square the number of spots we get, and that's our win. Here $f(x)$ is $x^2$ – a convex function. $\mathbb{E}[f(x)] = \frac{1+4+9+16+25+36}{6}=15\frac{1}{6}$. But $\mathbb{E}[x] = 3.5$, then $f(\mathbb{E}[x])=12.25$
#### • LONG ANSWER: a function $f(\dot)$ is <font color=7E87FC>convex</font> on an interval if for every $x$ and $y$ in that inverval 
#### $$f(\lambda x+(1-\lambda)y)≤\lambda f(x)+(1-\lambda)f(y)$$
#### for any $0≤\lambda≤1$. Grpahically: the line joining the points $(x,f(x))$ and $(y,f(y))$ is nowhere lower than the curve (<font color=7E87FC>concave</font> is the opposity, simply $-f$ is convex).
#### Suppose that a stock price $S$ is random and we want to consider the value of an option with payoff $P(S)$.  We could easily do $P(\mathbb{E}[S_T])$ at expiry. Alternatively we could look at the various option payoffs and then calculate the expected payoff as $\mathbb{E}[P(S_T)]$. The latter makes more sense and is indeed the correct way to value options w.r.t. the <font color=7E87FC>risk-neutral</font> stock price. If the payoff is convex then 
#### $$\mathbb{E}[P(S_T)]≥P(\mathbb{E}[S_T])$$
#### How much is the left side greater than the right one? Use Taylor series approximation around the mean of $S$.
#### $$S = \overline{S}+\varepsilon$$ $$\overline{S}=\mathbb{E}[S], \mathbb{E}[\varepsilon]=0$$ $$\mathbb{E}[f(S)]=\mathbb{E}[f(\overline{S}+\varepsilon)] = \mathbb{E}[f(\overline{S})+\varepsilon f'(\overline{S})+\frac{1}{2}\varepsilon^2f''(\overline{S})+\ldots] \approx f(\overline{S})+\frac{1}{2}f''(\overline{S})\mathbb{E}[\varepsilon^2] = f(\mathbb{E}[S])+\frac{1}{2}f''(\mathbb{E}[S])\mathbb{E}[\varepsilon^2]$$
#### The last term is exactly the amount by which the left side exceeds the right one.
#### Convexity adds to the price of the derivative.

### <font color=7E87FC>2.24 What is Itô's Lemma?</font>
#### • SHORT ANSWER: Itô's lemma is a chain rule for stochastic processes.
#### • EXAMPLE: an obvious example is given by
#### $$dS = \mu S dt + \sigma S dW$$ $$d(\ln S) = (\mu - \frac{1}{2}\sigma^2)dt + \sigma dW$$
#### • LONG ANWER:
#### $$dy = a(y,t)dt+b(y,t)dW$$
#### $f(y,t)$ is differentiable w.r.t. $t$ and twice differentiable w.r.t. $y$, then $f$ satisfies the following SDE:
#### $$df = \begin{pmatrix} \frac{\partial f}{\partial t} + a(y,t)\frac{\partial f}{\partial y} + \frac{1}{2}b(y,t)^2 \frac{\partial^2 f}{\partial y^2}\end{pmatrix}dt + b(y,t)\frac{\partial f}{\partial y}dW$$
#### Itô's lemma is to stochastic variables what Taylor series to deterministic ones.
#### In Itô's calculus:
##### – $dWdW = dt$;
##### – terms that are $O(dt^{3/2})$ or smaller must be ignored – $dt^2, dW^3, dtdW$ etc are too small to keep.
#### Suppose we have a function $f(y_1,y_2,\ldots,y_n,t)$ of $n$ stochastic quantities and time such that
#### $$dy_i = a_i(y_1,y_2,\ldots,y_n,t)dt + b_i(y_1,y_2,\ldots,y_n,t)dW_1$$
#### where $n$ Wiener processes $dW_i$ have correlations $\rho_{ij}$, then
#### $$df=\begin{pmatrix} \frac{\partial f}{\partial t} + \sum_{i=1}^na_i\frac{\partial f}{\partial y_i}+\frac{1}{2}\sum_{i=1}^n\sum_{j=1}^n \rho_{ij}b_ib_j\frac{\partial^2f}{\partial y_i\partial y_j}\end{pmatrix}dt + \sum_{i=1}^n b_i \frac{\partial f}{\partial y_i}dW_i$$ $$dW_i^2 = dt, dW_idW_j = \rho_{ij}dt$$

### <font color=7E87FC>2.25 Why Does Risk-Neutral Valuation Work?</font>
#### • SHORT ANSWER: risk-neutral valuation means we can value options in terms of their expected payoffs, discounted from expiration to the present, assuming that they grow on average at the risk-free rate.
#### The real rate at which the underlyign grows on average doesn't affect the value. For any derivative, as long as we can hedge it dynamically and perfectly, the hedged portfolio loses its randomness and behaves like a bond.
#### • EXAMPLE: current stock price = 44.75, average growth is 15% per annum. Vol. = 22%, interest rate = 4%. We want to value a call option with a strike of 45, expiring in two months.
#### First of all, the 15% avg. growth doesn't make sense. We shall use 4% risk-free rate instead, then we calculate the option's payoff for each path. Then we PV it all to today, and calculate the expected value. 
#### • LONG ANSWER: risk-neutral valuation of derivatives exploits the perfect correlation between the changes in the value of an option and its underlying.
#### Some explanations of risk-neutral pricing:
##### 1. if we hedge correctly in a Black-Scholes world, then all the risk is elimintaed. If there's no risk then we shouldn't expect any compensation for risk. We can therefore work under a measure in which everything grows at the risk free interest rate;
##### 2. if the model for the asset is $dS = \mu S dt + \sigma SdW$, then the $\mu$s cancel in the derivation of the Black-Scholes equation;
##### 3. two measures are <font color=7E87FC>equivalent</font> if they have the same sets of zero probability. A price is non-arbitrageable in the real world if and only if it's non-arbitrageable in the risk-neutral world. The risk-neutral price is ALWAYS non-arbitrageable.
##### Risk-neutral pricing only works under assumptions of continuous hedging, zero transaction costs, continuous asset paths, etc.

### <font color=7E87FC>2.26 What is Girsanov's Theorem, and Why is it Important in Finance?</font>
#### • SHORT ANSWER: it's the formal concept underlying the change of measure from the real world to the risk-neutral world. We can change from a Brownian motion with one drift to a BM with another.
#### • EXAMPLE: the classical example is:
#### $$dS = \mu S dt + \sigma S dW_t$$ $$dS = rSdt + \sigma S d\widetilde{W}_t$$
#### • LONG ANSWER: let $W_t$ be a Brownian motion with measure $\mathbb{P}$ and sample space $\Omega$. If $\gamma_t$ is a previsible (only depends on the previous history) process satisfying the constraint $\mathbb{E}_{\mathbb{P}}[\exp(\frac{1}{2}\int_0^T\gamma_t^2)]<\infty$ then there exists an equivalent measure $\mathbb{Q}$ on $\Omega$ such that 
#### $$\widetilde{W}_t = W_t + \int_0^t\gamma_sds$$
#### is a Brownian motion. In the differential form:
#### $$d\widetilde{W}_t = dW_t + \gamma_tdt$$
#### For example, we want to derive the valuation PDE for options under stochastic volatility. The stock price follows the real-world processes, $\mathbb{P}$,
#### $$dS = \mu S dt + \sigma S dW_1$$ $$d\sigma = a(S,\sigma,t)dt + b(S,\sigma,t)dW_2$$ $$dW_1dW_2 = \rho(S,\sigma,t)dt$$
#### Using Girsanov's theorem, we can get the governing equation in three steps:
#### 1. Under a pricing measure $\mathbb{Q}$, Girsanov plus the fact that $S$ is traded implies that ($\lambda$ = market price of volatility risk)
#### $$dW_1 = d\widetilde{W}_1-\frac{\mu-r}{\sigma}dt$$ $$dW_2 = d\widetilde{W_2}-\lambda(S,\sigma,t)dt;$$
#### 2. Apply Itô's formula to the discounted option price $V(S,\sigma,t)=e^{-r(T-t)}F(S,\sigma,t)$, expanding under $\mathbb{Q}$, using the formulæ for $dS$ and $dV$ obtained from the Girsanov transformation;
#### 3. Since the option is traded, the coefficient of the $dt$ term in its Itô expansion must also be zer; this yields the relevant equation.

### <font color=7E87FC>2.27 What are the Greeks?</font>
#### • SHORT ANSWER: the "greeks" are the sensitivities of derivatives prices to underlyings, variables and parameters.
#### • LONG ANSWER:
#### <font color=7E87FC>Delta $\Delta$</font>: $$\Delta = \frac{\partial V}{\partial S}$$
#### When shorting (in case of long option), remember about the cost of borrowing!
#### <font color=7E87FC>Gamma $\Gamma$</font>:
#### $$\Gamma = \frac{\partial^2V}{\partial S^2}$$
#### Because of large costs and because one wants to reduce exposure to model error, it's natural to try to minimize the need to rebalance too frequently. This can be achieved by buying or selling more OPTIONS!
#### <font color=7E87FC>Theta $\Theta$</font>:
#### $$\Theta = \frac{\partial V}{\partial t}$$
#### <font color=7E87FC>Speed</font> – the rate of change of the gamma with respect to the stock price:
#### $$\text{Speed} = \frac{\partial^3 V}{\partial S^3}$$
#### <font color=7E87FC>Vega/zeta/kappa</font>:
#### $$\text{Vega} = \frac{\partial V}{\partial \sigma}$$
#### Quite confusing, because it's a derivative w.r.t. a parameter, not a variable! That's an example of <font color=7E87FC>bastard greeks</font>. Call and put values have values that are monotonic in the volatility: increase the volatility, and the price of the option goes up.
#### <font color=7E87FC>Rho $\rho$</font>:
#### $$\rho = \frac{\partial V}{\partial r}$$
#### In practice one often uses a whole term structure of interest rate (time-dependent $r(t)$). Rho would then be the sensitivity to the level of the rates assuming a parallel shift in rates at all times.
#### <font color=7E87FC>Charm</font> – sensitivity of delta to time:
#### $$\text{Charm}=\frac{\partial^2V}{\partial S \partial t}$$
#### <font color=7E87FC>Colour</font> – rate of change of gamma with time:
#### $$\text{Colour}=\frac{\partial^3V}{\partial S^2 \partial t}$$
#### <font color=7E87FC>Vanna</font> – sensitivity of delta to volatility:
#### $$\text{Vanna} = \frac{\partial^2 V}{\partial S \partial \sigma}$$
#### When $\Gamma$ is small, forget about Vanna.
#### <font color=7E87FC>Vomma or Volga</font> – second derivative of the option value w.r.t. volatility.
#### $$\text{Vomma or Volga} = \frac{\partial^2V}{\partial \sigma^2}$$
#### Because of Jensen’s Inequality, if volatility is stochastic the Vomma/Volga measures convexity due to random volatility and so gives you an idea of how much to add (or subtract) from an option’s value.
#### <font color=7E87FC>Shadow greeks</font> – movement of several factors at once.

### <font color=7E87FC>2.29 What are the Forward and Backward Equations?</font>
#### • SHORT ANSWER: forward and backward equations refer to the differential equations governing the transition probability density function for a stochastic process.
#### • EXAMPLE: exchange rate is currently 1.88. What is the probability that it will be over 2 by this time next year?
#### • LONG ANSWER: we have a r.v. $y$ governed by one-factor SDE:
#### $$dy = A(y,t)dt + B(y,t)dW$$
#### $A,B$ – arbitrary functions of $y$ and $t$.
#### The <font color=7E87FC>transition PDF</font> $p(y,t;y',t')$ is the function of four variables defined by 
#### $$\text{Prob}(a<y<b\text{ at time }t'|y\text{ at time }t) = \int_a^b p(y,t;y',t')dy'$$
#### <font color=7E87FC>The forward equation</font> (Fokker-Planck, forward Kolmogorov equation):
#### $$\frac{\partial p}{\partial t'} = \frac{1}{2}\frac{\partial^2}{\partial y'^2}(B(y',t')^2p)-\frac{\partial}{\partial y'}(A(y',t')p)$$
#### This forward parabolic partial differential equation requires initial conditions at time $t$ and to be solved for $t' > t$.
#### <font color=7E87FC>The backward equation</font> (backward Kolmogorov equation):
#### $$\frac{\partial p}{\partial t}+\frac{1}{2}B(y,t)^2\frac{\partial^2p}{\partial y^2}+A(y,t)\frac{\partial p}{\partial y}=0$$
#### This must be solved backwards in t with specified final data. For example, if we wish to calculate the expected value of some function $F(S)$ at time $T$ we must solve this equation for the function $p(S,t)$ with
#### $$p(S,T) = F(S).$$

### <font color=7E87FC>2.30 What is the Black-Scholes Equation?</font>
#### • EXAMPLE:
$$\#frac{\partial V}{\partial t} + \frac{1}{2}\sigma^2S^2\frac{\partial^2V}{\partial S^2}+rS\frac{\partial V}{\partial S}-rV = 0$$
#### It's a backward equation, because we specify a final condition (option payoff at expiry) and solve backward in time to get the option value now. 
#### The equation contains four terms:
#### $\frac{\partial V}{\partial t}$ = time decay, how much the option value changes by if the stock price doesn't change;
#### $\frac{1}{2}\sigma^2S^2\frac{\partial^2V}{\partial S^2}$ = convexity term, how much a hedged position makes on average from stock moves;
#### $rS\frac{\partial V}{\partial S}$ = drift term allowing for the growth in the stock at the risk-free rate;
#### $-rV$ = the discounting term, since the payoff is received at expiration but we are valuing the option now.

### <font color=7E87FC>2.33 What is the Finite-Difference Method?</font>
#### • SHORT ANSWER: it's a way of approximating differential equations, in continuous variables, into difference equations, in discrete variables, so that they may be solved numerically. Technically, a def. of the delta of an option is:
#### $$\Delta = \frac{\partial V}{\partial S} = \underset{h\rightarrow 0}{\text{lim}} \frac{V(S+h,t)-V(S-h,t)}{2h}$$

#### Our task is to find numerically an approximation of the option values at each of the nodes on this grid.
#### An approximation:
#### $$\Delta = \frac{\partial V}{\partial S}\approx \frac{V(S+\delta S,t)-V(S-\delta S,t)}{2\delta S}$$
#### The central difference for the gamma is:
#### $$\Gamma \approx \frac{V(S+\delta S,t)-2V(S,t)+V(S-\delta S,t)}{\partial S^2}$$
#### We can write:
#### $$\Theta \approx \frac{V_i^k - V_i^{k-1}}{\delta t}$$ $$\Delta \approx \frac{V_{i+1}^k-V_{i-1}^k}{2\delta S}$$ $$\Gamma \approx \frac{V_{i+1}^k-2V_i^k + V_{i-1}^k}{\partial S^2}$$
#### We can plug it into the Black-Scholes equation ($S=i\delta S$):
#### $$\frac{V_i^k-V_i^{k-1}}{\delta t}+\frac{1}{2}\sigma^2i^2\delta S^2\frac{V_{i+1}^k-2V_i^k+V_{i-1}^k}{\delta S^2}+ri\delta S\frac{V_{i+1}^k-V_{i-1}^k}{2\delta S}-rV_i^k = 0$$
#### This can be rearranged to give $V_i^{k-1} $ in terms of $V_{i+1}^k, V_i^k, V_{i-1}^k$.

### <font color=7E87FC>2.34 What is a Poisson Process and What are its Uses in Finance?</font>
#### • SHORT ANSWER: it's a model for a discountinuous r.v. TIme is continuous, but the variable is discrete, it can represent a "jump" in a quantity or the occurence of an "event" (e.g. bankruptcy, default).
#### • LONG ANSWER: a Poisson process $dq$ is defined by the limit as $dt\rightarrow 0$ of:
#### $$dq = \begin{cases} 0, & \text{with probability } 1-\lambda dt, \\ 1, &\text{with probability } \lambda dt.\end{cases}$$
#### Therefore, there is a prob. $\lambda dt$ of a jump in $q$ in the time step $dt$. $\lambda$ = <font color=7E87FC>intensity</font> (parameter), possibly time dependent, or dep. on other vars in a model, that represents the likelihood of the jump.
#### When a moden has both $dW$ terms and a Poisson $dq$, it's called <font color=7E87FC>jump-diffusion model</font>.
#### General Poisson formula:
#### $$P(q(t)=n)=e^{-\lambda t}\frac{(\lambda t)^n}{n!}$$
#### E.g. probability of having $n$ buses arrived (or $n$ asset price jumps) in a time $t$.

### <font color=7E87FC>2.35 What is a Jump-Diffusion Model and How does it Affect Option Values?</font>
#### • SHORT ANSWER: $dW$ + $dq$. As we increase the frequency of the jumpts (all other params remain the same), the values of calls and puts increase.
#### • LONG ANSWER: the larger the $\lambda$, the more frequent the jumps. The prob. of a jump scales with the time period over which the jump may happen, $dt$. 
#### Combined with the Brownian motion, it's often taken to be:
#### $$dS = \mu S dt + \sigma S dW + (J-1)Sdq$$
#### $(J-1)$ = jump size.
#### Jump processes can be good for representing real-life phenomenon of discontinuity in vars and capturing fat tails seen in returns data.
#### Robert Merton was the first one to propose jump-diffusion models:
#### $$\frac{\partial V}{\partial t}+\frac{1}{2}\sigma^2 S^2 \frac{\partial^2V}{\partial S^2} + rS\frac{\partial S}{\partial S}-rV + \lambda \mathbb{E}[V(JS,t)-V(S,t)]-\lambda \frac{\partial V}{\partial S}S\mathbb{E}[J-1]=0$$
#### If the $J$ is lognorm. distributed, than the solution to this PDE is simple. The price of a European non-path-dependent option can be written as:
#### $$\sum_{n=0}^{\infty}\frac{1}{n!}e^{-\lambda'(T-t)}(\lambda'(T-t))^nV_{BS}(S,t;\sigma_n,r_n)$$ $$\lambda' = \lambda(1+k)$$ $$k=\mathbb{E}[J-1]$$ $$\sigma_{\Pi}^2 = \sigma^2 + \frac{n\sigma'^2}{T-t}$$ $$r_n = r-\lambda k + \frac{n\ln(1+k)}{T-t}$$
#### In other words, it's the sum of individual Black-Scholes values each of which assumes that there have been $n$ jumps, and they are weighted according to the probability that there will have been $n$ jumps before expiry.

### <font color=7E87FC>2.36 What is Meant by "Complete" and "Incomplete" Markets?</font>
#### • SHORT ANSWER: a complete market is one in which a derivative product can be artificially built from more basic instruments (cash, underlying). This involves dynamic rebalancing to replicate the derivative. An incomplete market is the one where you can't do that. 
#### • LONG ANSWER: more mathematically, complete market is one for which there exist the same number of linearly independent securities as there are sates of the world in the future.
#### For example, the binomial world with two possible sates with cash and stock -> complete market. HOWEVER, if the model was trinomial (up, same, down), then we would still need TWO instruments – dynamic hedging.
#### Recall that real world probabilities don't matter here. That's why people can disagree on the probs. of a stock rising or falling but still agree on the value of an option, as long as they share the same view on the stock's volatility.
#### Market is also complete when there exists a unique martingale measure.
#### <font color=7E87FC>In practice</font> all markets are incomplete because of real-world effects that violate the assumptions of the simple models. E.g. we don't know the future volatility (it's not deterministic), thus markets are incomplete (more states of the world than there are linearly independent secuirities). The same happens when the underlying follows a jump-diffusion process.

```python
def mjd_paths(S, T, r, sigma, lam, m, v, n_steps, n_paths):
    size = (n_steps, n_paths)
    dt = T/n_steps
    time = np.zeros(n_steps)
    poi_rv = np.multiply(np.random.poisson(lam*dt, size=size),
                        np.random.normal(m,v,size=size)).cumsum(axis=0)
    gbm = np.cumsum(((r-sigma**2/2-lam*(m+v**2/2))*dt + sigma*np.sqrt(dt)*\
                    np.random.normal(size=size)), axis=0)
    
    fig, ax = plt.subplots(figsize=(12,4), dpi=200)
    ax.plot(np.exp(gbm+poi_rv)*S, lw=.9)
    ax.margins(x=0.01)
    plt.show()
    return np.exp(gbm+poi_rv)*S
```

### <font color=7E87FC>2.40 What is GARCH?</font>
#### • SHORT ANSWER: stands for Generalized Auto Regressive Conditional Heteroskedasticity – econometric nodel for time-dependent variance, and hance vol. of stock price returns. It represents current variance in terms of past variance(s).
#### • EXAMPLE: simplest member of the GARCH family is GARCH(1,1), in which the variance $v_n$ of stock returns at time step $n$ is modelled by:
#### $$v_n = (1-\alpha-\beta)w_0+\beta v_{n-1}+\alpha v_{n-1}B_{n-1}^2$$
#### where $w_0$ is the long-term variance, $\alpha+\beta<1, \alpha > 0, \beta > 0$ are positive parameters, $B_n$ are independent Brownian increments. The latest variance $v_n$ can be thought of as a weighted average of the most recent variance, the latest square of returns, and the long-term average.
#### • LONG ANSWER: if we want to estimate vol. for $N$ days ahead, we must use $N$ previous days. That's a <font color=7E87FC>moving window</font>. This leads to artificially inflated volatility estimates for a while. 
#### Alternative for moving window is <font color=7E87FC>exponentially weighted MA (EWMA)</font>:
#### $$v_n = \beta v_{n-1}+(1-\beta)R_{n-1}^2$$
#### where $\beta$ is a parameter between 0 and 1, and the $R$’s are the returns, suitably normalized with the time step.
#### Here the latest variance is a weighted average between the previous variance and the latest square of returns. The higher the $\beta$, the more it depends on the distant returns data, and less on the recent. The problem is that it doesn't include term structure – forecast for tommorow or 1y ahead is just today's variance.
#### Finally, the third model is:
#### $$v_n = (1-\alpha-\beta)w_0 + \beta v_{n-1}+\alpha R_{n-1}^2$$
#### This is the GARCH(1,1) model, which adds a constant long-term variance to the EWMA. Expected variance $k$ steps in the future behaves like:
#### $$\mathbb{E}[v_{n+k}] = w_0 + (v_n-w_0)(\alpha+\beta)^n$$
#### Since $\alpha+\beta < 1$, this is exponentially decay of the average to its mean. More realistic than EWMA model.
#### <font color=7E87FC>General GARCH(p,q)</font> with $p$ past variances and $q$ past returns in the estimate:
#### $$v_n = \begin{pmatrix}1-\sum_{i=1}^q \alpha_i - \sum_{i=1}^p \beta_i\end{pmatrix}w_0 + \sum_{i=1}^p \beta_iv_{n-1} + \sum_{i=1}^q \alpha_i R_{n-i}^2$$
#### Parameters are usually estimated by <font color=7E87FC>MLE</font> applied to the (log)likelihood function.
#### <font color=7E87FC>Family members</font>:
#### – <font color=7E87FC>NGARCH</font> (non-linear GARCH), where $\gamma$ permits correlation between stock and volatility processes: 
#### $$v_n = (1-\alpha - \beta)w_0 + \beta v_{n-1} + \alpha (R_{n-1}-\gamma \sqrt{v_{n-1}})^2;$$
#### – <font color=7E87FC>AGARCH</font> (absolute GARCH) – volatility (not the variance) is linear in the absolute value of returns, instead of square of returns;
#### – <font color=7E87FC>EGARCH</font> (exponential GARCH) – models the logarithm of the variance, negative shocks can have a bigger impact on volatility than positive shocks;
#### – <font color=7E87FC>REGARCH</font> (range-based exponential GARCH) – models th elow to high range of asset prices over a 'day';
#### – <font color=7E87FC>IGARCH</font> (integrated GARCH);
#### – <font color=7E87FC>FIGARCH</font> (fractionally integrated GARCH) – uses fractional differencing lag operator applied to the variance, this models has the long memore, slow decay of volatility as seen in practice;
#### – <font color=7E87FC>FIEGARCH</font> (fractionally integrated exponential GARCH) – models the logarithm of variance and again has the long memore, slow decay of volatility;
#### – <font color=7E87FC>TGARCH</font> (threshold GARCH) – includes the extra term that kicks in when the sock is negative, gives a realistic asymmetry to the vol. model;
#### – <font color=7E87FC>PGARCH</font> (power GARCH) – variance is raised to a power other than zero (logarithm), one (AGARCH), or two;
#### – <font color=7E87FC>CGARCH</font> (component GARCH) – models variances as the sum of two or more components. In a two-component model one is used to capture short-term, another – the long-term effects of shocks.

### <font color=7E87FC>2.41 How Do I Dynamically Hedge?</font>
#### • SHORT ANSWER: it's continuous buying or selling of the underlying asset according to some algorithm, which eliminates the risk from an option position. The key point is what algorithm, formula to use, and how often (we can't hedge continuously).
#### • EXAMPLE: IV of a call option is 20%, but we think it's cheap and vol. is nearer 40%. Do we put 20% or 40% into the delta calculation?  The stock then moves, should we rebalance, incurring some inevitable transactions costs, or wait a bit longer while taking the risks of being unhedged?
#### • LONG ANSWER:
#### – <font color=7E87FC>What is the correct $\Delta$?</font>
#### Whatever we put in BSM (0.2 or 0.4), we won't lose money as long as we're right about the 40% and hedge continuously. However, choice of vol. will affect P&L. The profit will be ($\tilde{\sigma}$ = IV):
#### $$V(S,t;\sigma) - V(S,t;\tilde{\sigma})$$
#### When we hedge with IV, even though it's wrong compared with our forecast, we will still make money, but the profit each day is non-negative and smooth, so much nicer than when we hedge using forecast volatility. 
#### The downside – profit on the final day depends on the path taken by the asset. If the stock stays close to the stirke, we will make a lot of money. If the stock goes quiackly for ITM or OTM then the profit will be small. <font color=7E87FC>Hedging using IV</font> gives us a nice, smooth, monotonically increasing P&L, but at the cost of not knowing how much money we will make.
#### The profit each time step is ($\Gamma^i$ – gamma using IV):
#### $$\frac{1}{2}(\sigma^2 - \tilde{\sigma}^2)S^2\Gamma^i dt$$
#### As long as actual vol. is greater than IV, we will make money from this hedging strategy.
#### – <font color=7E87FC>How big is my hedging error?</font>
#### In practice we can't hedge continuously. Hedging discretely is quite easy to quantify. Our exposure becomes quadratic and depends on the gamma of our position. Let's use $\phi$ to denote a normally distributed r.v. $(0,1)$. Profit we'll make over a time step $\delta t$ due to the gamma is:
#### $$\frac{1}{2}\sigma^2S^2 \Gamma \delta t \phi^2$$
#### As long as we hedge discretely, our <font color=7E87FC>hedging error</font> (how much we make or lose money) is:
#### $$\frac{1}{2}\sigma^2S^2\Gamma \delta t(\phi^2 -1)$$
#### Some observations about hedging error:
##### – it's usually much bigger than interest received on the hedged option portfolio;
##### – on average it's zero;
##### – it's path dependent (the larger the gamma, the larger the hedging errors);
##### – total hedging error has std. dev. of $\sqrt{\delta t}$;
##### – hedging error is drawn from $\chi^2$ distribution;
##### – if we're long gamma we'll lose money approx. 68% of the time, but when we make money, it will be from the tails, and big enough to give a mean of zero. Short gamma we'll lose only 32% of the time, but they will be large losses;
##### – in practice $\phi$ isn't normally distributed: fat tails and high peaks (80% and 20% – see above).
#### – <font color=7E87FC>Can I optimize my hedge?</font>
#### Yes (see p. 343);
#### <font color=7E87FC>How much will transaction costs reduce my profit?</font>
#### The downside of frequent rehedging – transaction costs. As the stock changes by $\delta S$, the delta changes by $\delta S \Gamma$. If we assume that costs are proportional to the absolute value of the amount of the underlying traded $\kappa$, then the expected cost each $\delta t$ will be:
#### $$\kappa \sigma S^2 \sqrt{\delta t}\sqrt{\frac{2}{\pi}}|\Gamma|,$$
#### where $\sqrt{\frac{2}{\pi}}$ appears because we have to take the $\mathbb{E}[\cdot]$ of the absolute value of a normal r.v.
#### We can change the BS-PDE:
#### $$\frac{\partial V}{\partial t}+\frac{1}{2}\sigma^2S^2\frac{\partial V}{\partial S^2}+rS\frac{\partial V}{\partial S}-rV-\kappa\sigma S^2\sqrt{\frac{2}{\pi \delta t}}|\Gamma| = 0$$
#### This equation is non-linear: long calls will be less than the usual BSM value, and a short call – higher. The long position is worth less because we have to allow for the cost of hedging. The short is even more of a liability because of costs.

### <font color=7E87FC>2.44 What is Bootstrapping using Discount Factors?</font>
#### • SHORT ANSWER: it means building up a forward interest-rate curve that is consistent with the market prices of common FI-instruments such as bonds and swaps. The resulting curve can be used to value other instruments such as bonds that are not traded.
#### • EXAMPLE: if we know prices of bonds with maturities 1y, 2y, 3y, 5y, and we are aksed to value a 4y-bond, than we'll use boostrapping.
#### • LONG ANSWER: let's say we live in a world where rates are completely deterministic with no randomness at all. The spot rate is the one we receive from one instant to the next. In this deterministic interest-rate world this spot rate can be written as $r(t)$. If we knew this function, we would be able to value fixed-coupon bonds of all maturities by using the discount factor
#### $$\exp\begin{pmatrix} -\int_t^T r(\tau)d\tau\end{pmatrix}$$
#### to present value a payment at time $T$ to today's $t$.
#### Unfortunately we don't know what's the function, but we know some constraints on this $r(t)$ by looking at market prices of various FI-instruments.
#### Simple example: we know a zero-coupon bond with principal 100 USD maturing in 1y, and it's worth 95 USD:
#### $$\exp\begin{pmatrix}-\int_t^{t+1}r(\tau)d\tau\end{pmatrix} = 0.95$$
#### And similar 2y zero-coupon bond is worth 92 USD:
#### $$\exp\begin{pmatrix}-\int_t^{t+1}r(\tau)d\tau\end{pmatrix} = 0.92$$
#### Usually simple linear instruments are used for bootstrapping – bonds, swaps (they can be decomposed into a portfolio of bonds). Instruments such as caps and floors include optionality, which isn't linear, thus not used for bootstrapping.
#### Imposed conditions on $r(t)$:
##### – forward rates should be positive, or there will be arbitrage opportunities;
##### – forward rates should be continuous;
##### – perhaps the curve should also be smooth.
#### Even with the abovementioned points, the forward curve isn't uniqely defined.
#### <font color=7E87FC>Interpolation techniques</font>: linear in discount factors, in spot rates, in the log. of rates, piecewise lienar continuous forwards, cubic splines, Bessel cubic spline, monotone-preserving cubic spline, quadratic spline.
#### Because of liquidity, deposit rates are used in the very short term, bonds and FRAs – for the medium term, and swaps for the longer end of the forward curve.
#### Because the bootstrapped forward curve is assumed to come from deterministic rates it is dangerous to use it to price instruments with convexity since such instruments require a model for randomness, as explained by Jensen’s Inequality.

### <font color=7E87FC>2.54 Why is the Lognormal Distribution Important?</font>
#### • SHORT ANSWER: LN for assets themselves, N for returns.
#### • EXAMPLE: the famous $ds = \mu S dt + \sigma S dW$ results in a lognormal distribution, proveded $\mu, \sigma$ are independent of $S$.
#### • LONG ANSWER: the PDF of lognorm. distribution is:
#### $$\frac{1}{\sqrt{2\pi}bx}\exp\begin{pmatrix}-\frac{1}{2b^2}(\ln(x)-a)^2\end{pmatrix}, x≥0$$
#### where $a$ and $b>0$ represent the location and scale. The distrib. is scaled to the right, extending to infinity, and bounded below by zero.

### <font color=7E87FC>2.55 What are Copulas and How are they Used in Quantitative Finance?</font>
#### • SHORT ANSWER: copulas (from the Latin = "join") are used to model joint distribution of multiple underlyings. 
#### • EXAMPLE: say we have a basket of uncorrelated stocks, pretty much independent from each other. But on the day when the market moves dramatically they all moved together. Such behaviour can be modelled by copulas.
#### • LONG ANSWER: take $N$ uniformly distributed r.vs. $U_1, U_2, \ldots, U_N$ each defined on $[0,1]$. The copula function is def. as
#### $$C(u_1, u_2, \ldots, u_N) = \text{Prob}(U_1≤u_1, U_2≤u_2, \ldots, U_N≤u_N)$$
#### Clearly we have 
#### $$C(u_1,u_2,\ldots,0,\ldots,u_N) = 0$$ $$C(1,1,\ldots,u_i,\ldots,1)=u_i$$
#### Let $x_1, x_2, \ldots, x_N$ be r.vs. with CDF (marginal distributions) of $F_1(x_1), F_2(x_2), \ldots, F_N(x_N)$. Combine the $F$s with the copula function:
#### $$C(F_1(x_1),F_2(x_2), \ldots, F_N(x_N)) = F(x_1,x_2,\ldots,x_N)$$
#### The function $F(x_1,x_2,\ldots,x_N)$ is the same as
#### $$\text{Prob}(X_1≤x_1,X_2≤x_2,\ldots,X_N≤x_N)$$

### <font color=7E87FC>2.56 What is Asymptotic Analysis and How is it Used in Financial Modelling?</font>
#### • SHORT ANSWER: it's about exploiting a large or small parameter in a problem to find simple(r) equations or even solutions.
#### • EXAMPLE: the SABR model is a famous model for a forward rate and its vol. that exploits low vol. of vol. in order for closed-form solutions for option prices to be found.
#### • LONG ANSWER: outside finance asymptotic analysis is extremely common and useful.
#### We're looking at what happnes when a parameter gets small (denoted by $\epsilon$), and large parameters are $1/\epsilon$.
#### We say that <font color=7E87FC>$f(\epsilon) = O(g(\epsilon))$</font> as $\epsilon \rightarrow 0$ if $\underset{\epsilon \rightarrow 0}{\lim} \frac{f(\epsilon)}{g(\epsilon)}$ is finite.
#### We say that <font color=7E87FC>$f(\epsilon) = o(g(\epsilon))$</font> as $\epsilon \rightarrow 0$ if $\underset{\epsilon \rightarrow 0}{\lim} \frac{f(\epsilon)}{g(\epsilon)}\rightarrow 0$.
#### We say that <font color=7E87FC>$f(\epsilon) \sim g(\epsilon)$</font> as $\epsilon \rightarrow 0$ if $\underset{\epsilon\rightarrow 0}{\lim}\frac{f(\epsilon)}{g(\epsilon)}=1$.
#### There are several examples of asymptotic analysis in finance:
##### – transaction costs (usually a small percentage of trade);
##### – SABR (vol. of vol. is quite small);
##### – fast drift and high vol. in stoch. vol. models.

### <font color=7E87FC>2.57 What is a Free-Boundary Problem and What is the Optimal-Stopping Time for an American Option?</font>
#### • SHORT ANSWER: a boundary-value problem is typically a differential equation with specified solution on some domain. A free-boundary problem is one for which that boundary is also to be found as part of the solution. When to exercise an American option is an example of a free-boundary problem, the boundary representing th etime and place at which to exercise – optimal-stopping problem.
#### • LONG ANSWER: in a boundary-value problem the specification of the behaviour of the solution on some domain is to show that it has a unique solution; we must specify just the right type of conditions. In case of derivatives the boundary condition in time will be the final payoff – final condition. We must also specify two conditions in the asset space (e.g. put has zero value at infinite stock price, and is the discounted strike at zero stock price).
#### The American put must be exercised when the stock falls sufficiently low, but what is 'sufficient' here? 
##### – the option value must never fall below the payoff (or there would be arbitrage opportunities);
##### – we must exercise so as to give the option its highest value.
#### We must have option value and the payoff function continuous as functions of the underlying, and so does the delta function! This is called the <font color=7E87FC>smooth-pasting condition</font> (smooth = function and its first derivative are continuous).
#### Free-boundary problems are <font color=7E87FC>nonlinear</font>: we can't add two together to get another solution (e.g. American straddle is not the same as the sum of the American call and the American put).
#### <font color=7E87FC>The option is valued</font> by maximizing the value from the point of view of the delta-hedging writer.
#### Similar instruments with free-boundary problems: convertible bonds, callability, shout options, choosers.

### <font color=7E87FC>2.60 What are the Stupidest Things People have Said about Risk Neutrality?</font>
#### • EXAMPLE: saying "using risk-neutral pricing we replace $\mu$ with the risk-free $r$".
#### • LONG ANSWER: risk-neutral pricing means that we price as if traded contracts grow at the risk free rate, and that non-traded financial quantities have a growth that is the real growth adjusted for risk.
#### For this concept to work we need complete markets – enought traded quantities with which to hedge risk continuously, and we can't have jump conditions.
#### • CONFUSION OF REAL AND RISK NEUTRAL:
##### – the forward price = market's expected value of it in the future, which is WRONG! Under various assumptions there is a simple arbitrage between spot price and forward price that links them via the interest rate. So there is no information in the forward price about anyone’s expectations.
##### – the forward curve (of interest rates) is the market’s expected value of the spot interest rate at future times. Wrong. If any expectation is involved then it is a risk-neutral expectation. The forward curve contains information about the expected future spot interest rate, yes, but it also contains information about the market’s risk aversion.
##### – The delta of an option is the probability of it ending up in the money. Wrong for two reasons. One is that the probability of ending up in the money depends on the real probabilities, and the real growth rate, and that’s disappeared from the option value so it can’t be true. The second reason is that there’s a sign wrong! (If we have a call option then the probability you want is $N(d_2')$, where the prime ($'$) means use $\mu$ instead of $r$, and not $N(d_1)$).
